```python
class BinarySearchTree(object):
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None
    
    def insertNode(self, x):
        if x < self.key:
            if self.left:
                self.left.insertNode(x)
            else:
                tree = BinarySearchTree(x)
                self.left = tree
        elif x > self.key:
            if self.right:
                self.right.insertNode(x)
            else:
                tree = BinarySearchTree(x)
                self.right = tree
    
    def deleteNode(self,x):
        if self.findValue(x):
            if x < self.key:
                self.left = self.left.deleteNode(x)
                return self
            elif x > self.key:
                self.right = self.right.deleteNode(x)
                return self
            elif self.left and self.right:
                key = self.right.findMinimum().key
                self.key = key
                self.right = self.right.deleteNode(key)
                return self
            else:
                if self.left:
                    return self.left
                else:
                    return self.right
        else:
            return self
    
    def findValue(self, x):
        if x == self.key:
            return self
        elif x < self.key and self.left:
            return self.left.findValue(x)
        elif x > self.key and self.right:
            return self.right.findValue(x)
        else:
            return None
    
    def findMinimum(self):
        if self.left:
            return self.left.findMinimum()
        else:
            return self
    
    def findMaximum(self):
        tree = self
        if tree:
            while tree.right:
                tree = tree.right
        return tree
    
    # can i do this?
    def postOrder(self, l, root):
        if root:
            self.postOrder(l, root.left)
            self.postOrder(l, root.right)
            l.append(root.key)
        return l
        
    def sumNodes(self, root):
        l = []
        res = self.postOrder(l, root)
        currSum = 0
        for r in res:
            currSum = currSum + r
        return currSum
    
    def countNodes(self, root):
        l = []
        return len(self.postOrder(l, root))
    
    def printTree(self, root):
        if root:
            # post order
            self.printTree(root.left)
            self.printTree(root.right)
            print(root.key, " ")
    
    def measureHeight(self, root):
        if root:
            lHeight = self.measureHeight(root.left)
            rHeight = self.measureHeight(root.right)
            theHeight = max(lHeight, rHeight)
            theHeight = theHeight + 1
        else:
            theHeight = 0
        return theHeight
        
    
    def replaceNodeData(self, originRootValue, newRootValue):
        self.deleteNode(originRootValue)
        self.insertNode(newRootValue)
```
